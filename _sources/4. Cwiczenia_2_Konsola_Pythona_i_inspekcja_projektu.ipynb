### Zadanie 5.3

Uruchom ponizszy kod i zaobserwuj roznice w czasie wykonania:

```python
import time
from qgis.core import QgsFeatureRequest

layer = QgsProject.instance().mapLayersByName('kraje')[0]  # zmien nazwe
POLE = 'continent'
WARTOSC = 'Europe'

# Sposob 1: filtrowanie w Pythonie
t0 = time.time()
wynik1 = [ob for ob in layer.getFeatures() if ob[POLE] == WARTOSC]
t1 = time.time()
print(f'Filtrowanie Python : {len(wynik1)} obiektow w {t1-t0:.4f} s')

# Sposob 2: filtrowanie przez QgsFeatureRequest
t0 = time.time()
req = QgsFeatureRequest().setFilterExpression(f'"{POLE}" = \'{WARTOSC}\'')
wynik2 = list(layer.getFeatures(req))
t1 = time.time()
print(f'QgsFeatureRequest  : {len(wynik2)} obiektow w {t1-t0:.4f} s')
```
