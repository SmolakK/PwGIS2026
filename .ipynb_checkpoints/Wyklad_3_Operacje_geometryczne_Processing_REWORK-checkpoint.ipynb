## 1. Klasa QgsGeometry - przeglad metod

Klasa `QgsGeometry` reprezentuje geometrie pojedynczego obiektu wektorowego.
Jest to centralny obiekt przy wszelkich operacjach przestrzennych w PyQGIS.

### 1.1 Tworzenie geometrii

Geometrie mozna tworzyc na kilka sposobow:

```python
from qgis.core import QgsGeometry, QgsPointXY, QgsPoint, QgsWkbTypes

# Punkt
geom_punkt = QgsGeometry.fromPointXY(QgsPointXY(21.01, 52.23))

# Linia (lista punktow)
geom_linia = QgsGeometry.fromPolylineXY([
    QgsPointXY(17.0, 51.1),
    QgsPointXY(18.0, 51.5),
    QgsPointXY(19.0, 51.2),
])

# Polygon (lista pierscieni; pierwszy to obwod zewnetrzny)
geom_polygon = QgsGeometry.fromPolygonXY([[
    QgsPointXY(17.0, 51.0),
    QgsPointXY(18.0, 51.0),
    QgsPointXY(18.0, 52.0),
    QgsPointXY(17.0, 52.0),
    QgsPointXY(17.0, 51.0),  # zamkniecie pierscienia
]])

# Tworzenie z WKT (Well-Known Text)
geom_wkt = QgsGeometry.fromWkt('POINT(21.01 52.23)')
geom_wkt2 = QgsGeometry.fromWkt('POLYGON((17 51, 18 51, 18 52, 17 52, 17 51))')
```

### 1.2 Podstawowe wlasciwosci i metody pomiarowe

```python
# Typ geometrii
print(geom_polygon.type())                          # 2 = PolygonGeometry
print(QgsWkbTypes.displayString(geom_polygon.wkbType()))  # 'Polygon'

# Sprawdzenie poprawnosci
print(geom_polygon.isNull())    # True jesli geometria jest pusta
print(geom_polygon.isEmpty())   # True jesli nie zawiera wierzcholkow
print(geom_polygon.isGeosValid()) # True jesli geometria jest topologicznie poprawna

# Pomiary (jednostki CRS warstwy)
print(geom_polygon.area())      # pole powierzchni
print(geom_polygon.length())    # obwod (dla poligonu) lub dlugosc (dla linii)
print(geom_punkt.distance(geom_wkt))  # odleglosc miedzy geometriami

# Centroid i bounding box
print(geom_polygon.centroid().asPoint())  # srodek ciezkosci
print(geom_polygon.boundingBox())         # prostokat obwiedni

# Eksport do formatow tekstowych
print(geom_polygon.asWkt())     # WKT
print(geom_polygon.asJson())    # GeoJSON
```

### 1.3 Dostep do wspolrzednych

```python
# Punkt
punkt = geom_punkt.asPoint()     # zwraca QgsPointXY
print(punkt.x(), punkt.y())

# Linia
punkty_linii = geom_linia.asPolyline()  # lista QgsPointXY
for p in punkty_linii:
    print(p.x(), p.y())

# Polygon
pierscienie = geom_polygon.asPolygon()  # lista pierscieni, kazdy to lista QgsPointXY
obwod_zewnetrzny = pierscienie[0]
print(f'Liczba wierzcholkow: {len(obwod_zewnetrzny)}')
```


### Przykład 1 - dodanie punktu do mapy

````python
from qgis.core import (
    QgsVectorLayer,
    QgsFeature,
    QgsGeometry,
    QgsPointXY,
    QgsProject
)

# Tworzymy pustą warstwę punktową
layer = QgsVectorLayer("Point?crs=EPSG:4326", "Punkty_miasta", "memory")
prov = layer.dataProvider()

# Tworzymy punkt (np. Rynek Wrocław)
geom = QgsGeometry.fromPointXY(QgsPointXY(17.0385, 51.1090))

feat = QgsFeature()
feat.setGeometry(geom)

prov.addFeature(feat)
layer.updateExtents()

QgsProject.instance().addMapLayer(layer)
````

## Przykład 2 - linia

````python
from qgis.core import (
    QgsVectorLayer,
    QgsFeature,
    QgsGeometry,
    QgsPointXY,
    QgsProject
)

# Pusta warstwa liniowa
layer_linie = QgsVectorLayer("LineString?crs=EPSG:4326", "Trasy", "memory")
prov_linie = layer_linie.dataProvider()

# Geometria linii
geom_linia = QgsGeometry.fromPolylineXY([
    QgsPointXY(17.0385, 51.1090),
    QgsPointXY(17.0450, 51.1105),
    QgsPointXY(17.0520, 51.1120),
])

feat = QgsFeature()
feat.setGeometry(geom_linia)

prov_linie.addFeature(feat)
layer_linie.updateExtents()

QgsProject.instance().addMapLayer(layer_linie)

print("Długość linii:", geom_linia.length())
````

## 1b. Czytanie geometrii: co mozemy z niej wyciagnac?

Skoro mamy juz **punkt** i **linie**, to teraz pokazujemy studentom najwazniejsza rzecz:
**geometria to struktura danych**.  
Mozemy odczytac wspolrzedne, liczbe wierzcholkow, bounding box, centroid i odleglosci.

To jest bardzo dobry moment na live coding, bo wynik od razu widac i w konsoli, i na mapie.


````python

# [QGIS] Co mozemy wyciagnac z punktu i linii?
from qgis.core import QgsWkbTypes, QgsFeature

# 1. Punkt -> wspolrzedne
pt = geom.asPoint()
print('PUNKT')
print('  x =', pt.x())
print('  y =', pt.y())
print('  typ geometrii =', QgsWkbTypes.displayString(geom.wkbType()))

# 2. Linia -> lista punktow
pts = geom_linia.asPolyline()
print('
LINIA')
print('  liczba wierzcholkow =', len(pts))
print('  pierwszy punkt =', (pts[0].x(), pts[0].y()))
print('  ostatni punkt  =', (pts[-1].x(), pts[-1].y()))
print('  typ geometrii =', QgsWkbTypes.displayString(geom_linia.wkbType()))

# 3. Bounding box linii
bbox = geom_linia.boundingBox()
print('
BOUNDING BOX LINII')
print('  xmin =', bbox.xMinimum())
print('  xmax =', bbox.xMaximum())
print('  ymin =', bbox.yMinimum())
print('  ymax =', bbox.yMaximum())

# 4. Centroid linii
centroid = geom_linia.centroid()
print('
CENTROID LINII')
print('  punkt =', centroid.asPoint())

# 5. Odleglosc punkt -> linia
print('
ODLEGLOSC PUNKT-LINIA')
print('  distance =', geom.distance(geom_linia))

# 6. Dorzucmy centroid linii do warstwy punktowej, zeby bylo widac na mapie
feat_centroid = QgsFeature()
feat_centroid.setGeometry(centroid)
prov.addFeature(feat_centroid)
layer.updateExtents()
````

## 1c. Stolice i kraje w poblizu punktu

To jest sekcja, ktora zwykle dobrze przyciaga uwage:  
wybieramy **stolice Europy**, robimy wokol nich buffer i pytamy:

- jakie inne stolice leza w poblizu?
- jakie kraje przecina ten buffer?
- ktory kraj jest najblizej wskazanego punktu?

Mozna to pokazac jako mini-zapytanie przestrzenne na zywo.


````python

# [JUPYTER] Stolice Europy - kraje i stolice w poblizu punktu
# Dziala bez QGIS, dobra komorka do szybkiego pokazu w notatniku.

import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
import matplotlib.pyplot as plt

stolice = pd.DataFrame([
    ('Warszawa', 'Poland', 21.0122, 52.2297),
    ('Berlin', 'Germany', 13.4050, 52.5200),
    ('Prague', 'Czechia', 14.4378, 50.0755),
    ('Vienna', 'Austria', 16.3738, 48.2082),
    ('Bratislava', 'Slovakia', 17.1077, 48.1486),
    ('Budapest', 'Hungary', 19.0402, 47.4979),
    ('Vilnius', 'Lithuania', 25.2797, 54.6872),
    ('Copenhagen', 'Denmark', 12.5683, 55.6761),
    ('Amsterdam', 'Netherlands', 4.9041, 52.3676),
    ('Brussels', 'Belgium', 4.3517, 50.8503),
], columns=['capital', 'country', 'lon', 'lat'])

capitals = gpd.GeoDataFrame(
    stolice,
    geometry=gpd.points_from_xy(stolice.lon, stolice.lat),
    crs='EPSG:4326'
)

# pracujemy metrycznie
capitals_m = capitals.to_crs('EPSG:3035')
reference = capitals_m.loc[capitals_m['capital'] == 'Warszawa'].iloc[0].geometry
buffer_450km = reference.buffer(450_000)

nearby = capitals_m[capitals_m.geometry.within(buffer_450km)].copy()
nearby['dist_km'] = nearby.geometry.distance(reference) / 1000
nearby = nearby.sort_values('dist_km')

print('Stolice w promieniu 450 km od Warszawy:')
print(nearby[['capital', 'country', 'dist_km']].round({'dist_km': 1}).to_string(index=False))

fig, ax = plt.subplots(figsize=(7,7))
capitals_m.plot(ax=ax, color='lightgray', markersize=35)

gpd.GeoSeries([buffer_450km], crs='EPSG:3035').plot(ax=ax, facecolor='none', edgecolor='tab:red', linewidth=2)
nearby.plot(ax=ax, color='tab:blue', markersize=45)

for _, row in capitals_m.iterrows():
    ax.annotate(row['capital'], (row.geometry.x, row.geometry.y), fontsize=8, xytext=(3,3), textcoords='offset points')

ax.set_title('Stolice w poblizu Warszawy - buffer 450 km')
ax.set_axis_off()
plt.show()
````

````python

# [QGIS] To samo w QGIS: generujemy warstwe stolic i buffer wokol Warszawy
from qgis.core import QgsVectorLayer, QgsField, QgsFeature, QgsGeometry, QgsPointXY, QgsProject
from PyQt5.QtCore import QVariant

cap_layer = QgsVectorLayer('Point?crs=EPSG:4326&field=capital:string&field=country:string', 'Stolice_Europy', 'memory')
prov_cap = cap_layer.dataProvider()

capital_data = [
    ('Warszawa', 'Poland', 21.0122, 52.2297),
    ('Berlin', 'Germany', 13.4050, 52.5200),
    ('Prague', 'Czechia', 14.4378, 50.0755),
    ('Vienna', 'Austria', 16.3738, 48.2082),
    ('Bratislava', 'Slovakia', 17.1077, 48.1486),
    ('Budapest', 'Hungary', 19.0402, 47.4979),
    ('Vilnius', 'Lithuania', 25.2797, 54.6872),
]

feats = []
for capital, country, lon, lat in capital_data:
    f = QgsFeature(cap_layer.fields())
    f.setAttributes([capital, country])
    f.setGeometry(QgsGeometry.fromPointXY(QgsPointXY(lon, lat)))
    feats.append(f)

prov_cap.addFeatures(feats)
cap_layer.updateExtents()
QgsProject.instance().addMapLayer(cap_layer)

print('Dodano warstwe stolic. Teraz mozna:')
print('1) zreprojektowac warstwe do metrycznego CRS')
print('2) zrobic buffer wokol Warszawy')
print('3) sprawdzic, ktore stolice wpadaja do bufora')
````

## 2. Operacje geometryczne - buforowanie i upraszczanie

### 2.1 Buforowanie

Bufor (ang. *buffer*) to obszar otaczajacy geometrie w zadanej odleglosci.
W PyQGIS metoda `.buffer()` dziala na pojedynczej geometrii.

```python
geom_punkt = QgsGeometry.fromPointXY(QgsPointXY(21.01, 52.23))

# Bufor o promieniu 1.0 (jednostki CRS - tu stopnie)
# Drugi argument to liczba segmentow aproksymujacych luk okragly
bufor = geom_punkt.buffer(1.0, segments=32)

print(f'Typ bufora    : {QgsWkbTypes.displayString(bufor.wkbType())}')
print(f'Pole pow.     : {bufor.area():.4f}')
print(f'WKT (fragment): {bufor.asWkt()[:80]}...')
```

**Uwaga dotyczaca jednostek:** Metoda `.buffer()` przyjmuje odleglosc
w jednostkach CRS geometrii. Jesli CRS jest geograficzny (stopnie, EPSG:4326),
wartosc `1.0` oznacza 1 stopien geograficzny (~111 km na rowniku), a nie 1 metr.
Aby bufforowac w metrach, nalezy wczesniej reprojekcjonowac geometrie
do ukladu metrycznego (np. EPSG:2180 dla Polski).

### 2.2 Upraszczanie geometrii

Upraszczanie (ang. *simplification*) redukuje liczbe wierzcholkow geometrii
zachowujac ogolny ksztalt. Przydatne przy wizualizacji lub eksporcie danych.

```python
layer = iface.activeLayer()

for obiekt in layer.getFeatures():
    geom = obiekt.geometry()
    n_przed = geom.constGet().nCoordinates()

    # Tolerancja upraszczania w jednostkach CRS
    uproszczona = geom.simplify(0.01)
    n_po = uproszczona.constGet().nCoordinates()

    print(f'{obiekt["name"]:<20} wierzcholki: {n_przed:>6} -> {n_po:>6}')
    break  # tylko pierwszy obiekt dla demonstracji
```

### 2.3 Inne transformacje jednej geometrii

```python
geom = obiekt.geometry()

# Otoczka wypukla (convex hull)
hull = geom.convexHull()

# Centroid
centroid = geom.centroid()

# Punkt na powierzchni (gwarantowanie wewnatrz poligonu)
punkt_wew = geom.pointOnSurface()

# Prostokat obwiedni jako geometria
bbox_geom = QgsGeometry.fromRect(geom.boundingBox())

# Rozszerzenie (positive = outward, negative = inward)
offsetted = geom.offsetCurve(0.1, 8, 0, 5.0)  # dla linii
```


````python
# [JUPYTER] Demonstracja buforowania przez Shapely

try:
    from shapely.geometry import Point, Polygon
    import matplotlib.pyplot as plt
    import matplotlib.patches as mpatches

    fig, axes = plt.subplots(1, 3, figsize=(13, 4))

    # Bufor wokol punktu
    p = Point(0, 0)
    for ax, seg in zip(axes, [4, 8, 32]):
        buf = p.buffer(1.0, resolution=seg)
        x, y = buf.exterior.xy
        ax.fill(x, y, alpha=0.3, color='steelblue')
        ax.plot(0, 0, 'ro', markersize=6)
        ax.set_aspect('equal')
        ax.set_title(f'segments={seg} ({len(list(buf.exterior.coords))} wierzcholkow)')
        ax.grid(True, alpha=0.3)

    plt.suptitle('Bufor punktu - wplyw liczby segmentow', fontsize=12)
    plt.tight_layout()
    plt.show()

    # Rozne rodzaje buforow dla poligonu
    fig, axes = plt.subplots(1, 3, figsize=(13, 4))
    poly = Polygon([(0,0),(3,0),(3,2),(0,2)])

    for ax, (d, title) in zip(axes, [
        (0.3,  'Bufor zewnetrzny (+0.3)'),
        (-0.3, 'Bufor wewnetrzny (-0.3)'),
        (0.0,  'Otoczka wypukla'),
    ]):
        x0, y0 = poly.exterior.xy
        ax.fill(x0, y0, alpha=0.2, color='grey')
        ax.plot(x0, y0, 'grey', linewidth=1)

        if d == 0.0:
            result = poly.convex_hull
        else:
            result = poly.buffer(d)

        if hasattr(result, 'exterior'):
            xr, yr = result.exterior.xy
            ax.fill(xr, yr, alpha=0.4, color='steelblue')
            ax.plot(xr, yr, 'steelblue', linewidth=2)

        ax.set_aspect('equal')
        ax.set_title(title)
        ax.grid(True, alpha=0.3)

    plt.suptitle('Transformacje poligonu', fontsize=12)
    plt.tight_layout()
    plt.show()

except ImportError:
    print('Zainstaluj: pip install shapely matplotlib')
````

## 3. Operacje na zbiorach geometrii

Operacje na zbiorach (ang. *set operations*) dotycza relacji miedzy dwiema geometriami.
Wszystkie ponizsze metody sa dostepne na obiekcie `QgsGeometry`
i zwracaja nowy obiekt `QgsGeometry`.

### 3.1 Przeglad operacji

| Metoda | Opis | Analogia |
|---|---|---|
| `.intersection(other)` | Czesc wspolna dwoch geometrii | A i B |
| `.union(other)` | Suma dwoch geometrii | A lub B |
| `.difference(other)` | Czesc A bez czesci wspolnej z B | A bez B |
| `.symDifference(other)` | Czesci nienalezace do czesci wspolnej | A lub B, ale nie obydwa |
| `.combine(other)` | Polaczenie bez scalania (GeometryCollection) | A + B obok siebie |

### 3.2 Przyklad w PyQGIS

```python
from qgis.core import QgsGeometry

# Dwa nachodzace na siebie prostokatne poligony
A = QgsGeometry.fromWkt('POLYGON((0 0, 2 0, 2 2, 0 2, 0 0))')
B = QgsGeometry.fromWkt('POLYGON((1 1, 3 1, 3 3, 1 3, 1 1))')

przeciecie = A.intersection(B)
print(f'Przeciecie   : {przeciecie.asWkt()}')
print(f'Pole pow.    : {przeciecie.area():.2f}')

suma = A.union(B)
print(f'Suma         : {suma.asWkt()}')
print(f'Pole pow.    : {suma.area():.2f}')

roznica = A.difference(B)
print(f'Roznica A-B  : {roznica.asWkt()}')
print(f'Pole pow.    : {roznica.area():.2f}')

sym_diff = A.symDifference(B)
print(f'Sym. roznica : {sym_diff.asWkt()}')
print(f'Pole pow.    : {sym_diff.area():.2f}')
```

### 3.3 Sprawdzanie poprawnosci przed operacja

Operacje geometryczne moga sie nie powiescc jesli geometria jest niepoprawna
topologicznie (np. polygon z samoprzecieciem). Zawsze warto sprawdzic
i ewentualnie naprawic geometrie:

```python
geom = obiekt.geometry()

if not geom.isGeosValid():
    bledy = geom.validateGeometry()
    for blad in bledy:
        print(f'Blad: {blad.what()} w punkcie {blad.where()}')

    # Proba naprawy
    naprawiona = geom.makeValid()
    if naprawiona.isGeosValid():
        geom = naprawiona
        print('Geometria naprawiona.')
    else:
        print('Nie mozna naprawic geometrii.')
```


````python
# [JUPYTER] Wizualizacja operacji na zbiorach geometrii

try:
    from shapely.geometry import Polygon
    from shapely.ops import unary_union
    import matplotlib.pyplot as plt
    import matplotlib.patches as mpatches

    A = Polygon([(0,0),(2,0),(2,2),(0,2)])
    B = Polygon([(1,1),(3,1),(3,3),(1,3)])

    operacje = [
        ('A', A, 'lightgrey'),
        ('B', B, 'lightgrey'),
        ('Przeciecie (A i B)', A.intersection(B), 'steelblue'),
        ('Suma (A lub B)', A.union(B), 'steelblue'),
        ('Roznica (A bez B)', A.difference(B), 'steelblue'),
        ('Sym. roznica', A.symmetric_difference(B), 'steelblue'),
    ]

    fig, axes = plt.subplots(2, 3, figsize=(13, 8))
    axes = axes.flatten()

    for ax, (tytul, geom, kolor) in zip(axes, operacje):
        # Tlo - oryginalne poligony
        for g, c in [(A, '#ffcccc'), (B, '#ccccff')]:
            x, y = g.exterior.xy
            ax.fill(x, y, alpha=0.3, color=c)
            ax.plot(x, y, color='grey', linewidth=1, linestyle='--')

        # Wynik operacji
        if not geom.is_empty:
            geoms = [geom] if geom.geom_type != 'MultiPolygon' else list(geom.geoms)
            for g in geoms:
                if hasattr(g, 'exterior'):
                    x, y = g.exterior.xy
                    ax.fill(x, y, alpha=0.6, color=kolor)
                    ax.plot(x, y, color='navy', linewidth=2)

        ax.set_xlim(-0.3, 3.3)
        ax.set_ylim(-0.3, 3.3)
        ax.set_aspect('equal')
        ax.set_title(tytul, fontsize=10)
        area = geom.area if not geom.is_empty else 0
        ax.set_xlabel(f'Pole pow.: {area:.2f}', fontsize=9)
        ax.grid(True, alpha=0.3)

    plt.suptitle('Operacje na zbiorach geometrii', fontsize=13)
    plt.tight_layout()
    plt.show()

except ImportError:
    print('Zainstaluj: pip install shapely matplotlib')
````

## 4. Relacje przestrzenne miedzy geometriami

Relacje przestrzenne odpowiadaja na pytanie: jak dwie geometrie sa polozony
wzgledem siebie? Sa podstawa zapytan przestrzennych (ang. *spatial queries*).

### 4.1 Metody relacji w QgsGeometry

| Metoda | Pytanie | Zwraca |
|---|---|---|
| `.intersects(other)` | Czy geometrie sie przecinaja lub dotykaja? | `bool` |
| `.contains(other)` | Czy A zawiera B w calosci? | `bool` |
| `.within(other)` | Czy A jest calkowicie wewnatrz B? | `bool` |
| `.touches(other)` | Czy geometrie stykaja sie tylko na granicy? | `bool` |
| `.overlaps(other)` | Czy A i B czesciowo sie pokrywaja? | `bool` |
| `.disjoint(other)` | Czy geometrie nie maja nic wspolnego? | `bool` |
| `.distance(other)` | Jaka jest odleglosc miedzy geometriami? | `float` |

### 4.2 Przyklad - znajdowanie obiektow w zasięgu

```python
from qgis.core import QgsGeometry, QgsPointXY, QgsFeatureRequest

layer = iface.activeLayer()

# Punkt referencyjny - np. centrum Wroclawia
punkt_ref = QgsGeometry.fromPointXY(QgsPointXY(17.038, 51.108))

# Znajdz wszystkie obiekty przecinajace sie z buforem punktu
promien = 2.0  # 2 stopnie (~220 km)
bufor = punkt_ref.buffer(promien, 32)

print(f'Obiekty w promieniu {promien} stopni od Wroclawia:')
for obiekt in layer.getFeatures():
    geom = obiekt.geometry()
    if geom.intersects(bufor):
        print(f'  {obiekt["name"]}')
```

### 4.3 Wydajniejszy wariant przez QgsFeatureRequest

Powyzszy przyklad iteruje po wszystkich obiektach i sprawdza kazdy z osobna.
Dla duzych warstw lepiej uzyc `QgsFeatureRequest.setFilterRect()`,
ktory wstepnie odfiltrowuje obiekty przez indeks przestrzenny:

```python
from qgis.core import QgsFeatureRequest

# Ogranicz obszar poszukiwan do bounding box buforu
bbox = bufor.boundingBox()
req = QgsFeatureRequest().setFilterRect(bbox)

wynik = []
for obiekt in layer.getFeatures(req):
    # Dokladne sprawdzenie przeciecia (bbox moze dawac false positives)
    if obiekt.geometry().intersects(bufor):
        wynik.append(obiekt['name'])

print(f'Znaleziono {len(wynik)} obiektow: {wynik}')
```

Wzorzec: najpierw `setFilterRect()` jako szybki filtr wstepny,
potem dokladne sprawdzenie geometryczne jako filtr wtorny.
Dzieki temu unikamy pelnego skanu warstwy.


````python
# [JUPYTER] Demonstracja relacji przestrzennych przez Shapely

try:
    from shapely.geometry import Point, Polygon, LineString

    A = Polygon([(0,0),(4,0),(4,4),(0,4)])
    B = Polygon([(2,2),(6,2),(6,6),(2,6)])
    C = Polygon([(5,5),(8,5),(8,8),(5,8)])
    D = Polygon([(1,1),(3,1),(3,3),(1,3)])  # wewnatrz A
    p = Point(2, 2)                          # na granicy A i B

    print('Relacje miedzy geometriami:')
    print(f'  A.intersects(B) = {A.intersects(B)}')
    print(f'  A.intersects(C) = {A.intersects(C)}')
    print(f'  A.contains(D)   = {A.contains(D)}')
    print(f'  D.within(A)     = {D.within(A)}')
    print(f'  A.disjoint(C)   = {A.disjoint(C)}')
    print(f'  A.overlaps(B)   = {A.overlaps(B)}')
    print()
    print('Odleglosci:')
    print(f'  A.distance(B) = {A.distance(B):.2f} (przecinaja sie -> 0)')
    print(f'  A.distance(C) = {A.distance(C):.4f}')
    print(f'  B.distance(C) = {B.distance(C):.4f}')

except ImportError:
    print('Zainstaluj: pip install shapely')
````

## 5. Reprojekcja geometrii miedzy ukladami wspolrzednych

Reprojekcja to przeliczenie wspolrzednych geometrii z jednego ukladu CRS
do innego. Jest konieczna gdy chcemy:

- wykonywac pomiary w metrach (z danych w stopniach),
- laczyc warstwy z roznych zrodel o roznych ukladach,
- zapisac wyniki w okreslonym ukladzie krajowym.

### 5.1 Klasy transformacji w PyQGIS

Do reprojekcji uzywamy dwoch klas:

- `QgsCoordinateReferenceSystem` - reprezentuje uklad wspolrzednych
- `QgsCoordinateTransform` - wykonuje transformacje miedzy dwoma ukladami

```python
from qgis.core import (
    QgsCoordinateReferenceSystem,
    QgsCoordinateTransform,
    QgsProject
)

# Definicja ukladow
crs_wgs84 = QgsCoordinateReferenceSystem('EPSG:4326')
crs_pl1992 = QgsCoordinateReferenceSystem('EPSG:2180')

# Obiekt transformacji
transformacja = QgsCoordinateTransform(
    crs_wgs84,    # zrodlowy CRS
    crs_pl1992,   # docelowy CRS
    QgsProject.instance()
)
```

### 5.2 Reprojekcja pojedynczej geometrii

```python
geom = QgsGeometry.fromPointXY(QgsPointXY(17.038, 51.108))  # Wroclaw w WGS84

# transform() modyfikuje geometrie IN-PLACE i zwraca kod bledu
blad = geom.transform(transformacja)

if blad == QgsGeometry.Success:
    punkt = geom.asPoint()
    print(f'Wroclaw w EPSG:2180:')
    print(f'  X (Easting) : {punkt.x():.2f} m')
    print(f'  Y (Northing): {punkt.y():.2f} m')
else:
    print(f'Blad transformacji: {blad}')
```

**Uwaga:** Metoda `.transform()` modyfikuje geometrie w miejscu (*in-place*).
Jesli chcesz zachowac oryginal, zrob kopie przez `QgsGeometry(geom)` przed transformacja.

### 5.3 Reprojekcja calej warstwy w petli

```python
from qgis.core import (
    QgsCoordinateReferenceSystem, QgsCoordinateTransform, QgsProject
)

layer = iface.activeLayer()

crs_zrodlowy = layer.crs()
crs_docelowy = QgsCoordinateReferenceSystem('EPSG:2180')

transformacja = QgsCoordinateTransform(
    crs_zrodlowy,
    crs_docelowy,
    QgsProject.instance()
)

print(f'Reprojekcja z {crs_zrodlowy.authid()} do {crs_docelowy.authid()}')

for obiekt in layer.getFeatures():
    geom = QgsGeometry(obiekt.geometry())  # kopia geometrii
    geom.transform(transformacja)

    # Tu mozna np. obliczyc pole pow. w metrach kwadratowych
    if geom.type() == 2:  # polygon
        print(f'  {obiekt["name"]:<20} {geom.area() / 1e6:.0f} km2')
    break  # tylko pierwszy obiekt dla demonstracji
```

### 5.4 Pomiary metryczne z QgsDistanceArea

Alternatywa dla recznej reprojekcji jest klasa `QgsDistanceArea`,
ktora sama zajmuje sie przeliczeniami jednostek:

```python
from qgis.core import QgsDistanceArea, QgsProject

kalkulator = QgsDistanceArea()
kalkulator.setEllipsoid('WGS84')
kalkulator.setSourceCrs(
    layer.crs(),
    QgsProject.instance().transformContext()
)

for obiekt in layer.getFeatures():
    geom = obiekt.geometry()
    if geom.type() == 2:  # polygon
        pole_m2 = kalkulator.measureArea(geom)
        pole_km2 = kalkulator.convertAreaMeasurement(
            pole_m2, QgsUnitTypes.AreaSquareKilometers
        )
        print(f'{obiekt["name"]:<20} {pole_km2:.0f} km2')
    break
```


````python
# [JUPYTER] Demonstracja reprojekcji przez pyproj

try:
    from pyproj import Transformer

    # Transformer: z WGS84 do Polska CS92 (EPSG:2180)
    t = Transformer.from_crs('EPSG:4326', 'EPSG:2180', always_xy=True)

    miasta = [
        ('Wroclaw',  17.038, 51.108),
        ('Warszawa', 21.012, 52.230),
        ('Krakow',   19.945, 50.065),
        ('Gdansk',   18.646, 54.352),
        ('Poznan',   16.929, 52.407),
    ]

    print(f'{"Miasto":<12} {"lon (WGS84)":>12} {"lat (WGS84)":>12}  ->  '
          f'{"X (CS92)":>12} {"Y (CS92)":>12}')
    print('-' * 72)

    for nazwa, lon, lat in miasta:
        x, y = t.transform(lon, lat)
        print(f'{nazwa:<12} {lon:>12.4f} {lat:>12.4f}  ->  {x:>12.2f} {y:>12.2f}')

    # Odleglosc Wroclaw-Warszawa w metrach
    import math
    x1, y1 = t.transform(17.038, 51.108)
    x2, y2 = t.transform(21.012, 52.230)
    odleglosc = math.sqrt((x2-x1)**2 + (y2-y1)**2)
    print(f'\nOdleglosc Wroclaw-Warszawa: {odleglosc/1000:.1f} km')

except ImportError:
    print('Zainstaluj: pip install pyproj')
````

## 5b. Generowanie wlasnych danych: siatka, linie, ksztalty

To jest bardzo wdzieczna czesc live codingu, bo pokazuje, ze w GIS **nie musimy miec gotowego pliku**.
Mozemy wygenerowac dane programowo:

- siatke regularna,
- linie promieniste,
- proste ksztalty testowe,
- syntetyczne obiekty do dalszych analiz.

To przydaje sie np. do testow, symulacji, wizualizacji i automatycznego przygotowania danych.


````python

# [QGIS] Generowanie siatki kwadratow 5x5 jako warstwy polygonowej
from qgis.core import QgsVectorLayer, QgsFeature, QgsGeometry, QgsPointXY, QgsProject

cell_size = 1.0
n = 5
origin_x, origin_y = 0.0, 0.0

grid = QgsVectorLayer('Polygon?crs=EPSG:4326&field=id:integer&field=row:integer&field=col:integer', 'Siatka_5x5', 'memory')
prov_grid = grid.dataProvider()

features = []
idx = 1
for r in range(n):
    for c in range(n):
        x0 = origin_x + c * cell_size
        y0 = origin_y + r * cell_size
        ring = [
            QgsPointXY(x0, y0),
            QgsPointXY(x0 + cell_size, y0),
            QgsPointXY(x0 + cell_size, y0 + cell_size),
            QgsPointXY(x0, y0 + cell_size),
            QgsPointXY(x0, y0)
        ]
        f = QgsFeature(grid.fields())
        f.setAttributes([idx, r, c])
        f.setGeometry(QgsGeometry.fromPolygonXY([ring]))
        features.append(f)
        idx += 1

prov_grid.addFeatures(features)
grid.updateExtents()
QgsProject.instance().addMapLayer(grid)
print(f'Utworzono {len(features)} pol w siatce.')
````

````python

# [QGIS] Linie promieniste wychodzace z jednego punktu
import math
from qgis.core import QgsVectorLayer, QgsFeature, QgsGeometry, QgsPointXY, QgsProject

center = QgsPointXY(17.0385, 51.1090)
radius = 0.6
n_rays = 12

rays = QgsVectorLayer('LineString?crs=EPSG:4326&field=id:integer&field=angle:double', 'Promienie', 'memory')
prov_rays = rays.dataProvider()

features = []
for i in range(n_rays):
    angle = 2 * math.pi * i / n_rays
    x = center.x() + radius * math.cos(angle)
    y = center.y() + radius * math.sin(angle)
    geom_ray = QgsGeometry.fromPolylineXY([center, QgsPointXY(x, y)])

    f = QgsFeature(rays.fields())
    f.setAttributes([i + 1, round(math.degrees(angle), 1)])
    f.setGeometry(geom_ray)
    features.append(f)

prov_rays.addFeatures(features)
rays.updateExtents()
QgsProject.instance().addMapLayer(rays)
print(f'Utworzono {len(features)} linii promienistych.')
````

````python

# [JUPYTER] Wlasne ksztalty i operacje na nich
from shapely.geometry import Point, Polygon, LineString
import matplotlib.pyplot as plt
import numpy as np

circle = Point(0, 0).buffer(1.2, resolution=64)
star = Polygon([
    (0, 1.4), (0.3, 0.4), (1.3, 0.4), (0.5, -0.2),
    (0.8, -1.2), (0, -0.5), (-0.8, -1.2), (-0.5, -0.2),
    (-1.3, 0.4), (-0.3, 0.4), (0, 1.4)
])

inter = circle.intersection(star)
union = circle.union(star)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for ax, geom, title in zip(axes, [circle, star, inter], ['Okrag (buffer)', 'Gwiazda', 'Przeciecie']):
    x, y = geom.exterior.xy
    ax.fill(x, y, alpha=0.45)
    ax.plot(x, y)
    ax.set_title(title)
    ax.set_aspect('equal')
    ax.grid(alpha=0.25)
plt.tight_layout()
plt.show()
````

## 6c. Zaleznosci miedzy panstwami: sasiedzi i centroidy

To jest bardzo dobra sekcja na koniec wykladu przed bardziej technicznymi tematami.
Pokazuje, ze relacje przestrzenne mozna interpretowac jak **graf polaczen**:

- kraj A graniczy z krajem B,
- kraj B graniczy z krajem C,
- wiec C jest "sasiadem drugiego stopnia" kraju A.

Dodatkowo mozemy pokazac ciekawy przypadek geometryczny:
**centroid kraju nie zawsze lezy w samym kraju**.
Dzieje sie tak np. dla panstw o nietypowym ksztalcie lub wielowyspowych.


````python

# [JUPYTER] Sasiedzi pierwszego i drugiego stopnia
import geopandas as gpd

url = 'https://naturalearth.s3.amazonaws.com/110m_cultural/ne_110m_admin_0_countries.zip'
world = gpd.read_file(url)
eu = world[world['CONTINENT'] == 'Europe'].copy().reset_index(drop=True)
eu = eu[['ADMIN', 'geometry']].rename(columns={'ADMIN': 'country'})

# budujemy relacje sasiedztwa przez intersects/touches
neighbors = {country: set() for country in eu['country']}
for i, row_i in eu.iterrows():
    for j, row_j in eu.iterrows():
        if i >= j:
            continue
        if row_i.geometry.touches(row_j.geometry):
            neighbors[row_i['country']].add(row_j['country'])
            neighbors[row_j['country']].add(row_i['country'])

country = 'Poland'
first_degree = neighbors[country]
second_degree = set()
for n in first_degree:
    second_degree |= neighbors[n]
second_degree -= first_degree
second_degree.discard(country)

print(f'Sasiedzi pierwszego stopnia dla {country}:')
print(sorted(first_degree))
print(f'
Sasiedzi drugiego stopnia dla {country}:')
print(sorted(second_degree))
````

````python

# [JUPYTER] Kraje, ktorych centroid nie lezy w kraju
import geopandas as gpd

url = 'https://naturalearth.s3.amazonaws.com/110m_cultural/ne_110m_admin_0_countries.zip'
world = gpd.read_file(url)
world = world[['ADMIN', 'CONTINENT', 'geometry']].rename(columns={'ADMIN': 'country'})

world_eq = world.to_crs('EPSG:8857')  # Equal Earth - sensowny do globalnych centroidow
centroids = world_eq.geometry.centroid
inside = centroids.within(world_eq.geometry)

weird = world_eq.loc[~inside, ['country', 'CONTINENT']].copy()
weird['reason_hint'] = 'nietypowy ksztalt / wiele wysp / bardzo wydluzona geometria'

print('Przyklady krajow, gdzie centroid nie wpada do wlasnej geometrii:')
print(weird.head(15).to_string(index=False))
````

````python

# [JUPYTER] Wizualizacja jednego takiego przypadku - np. Norwegia lub Grecja, jesli wystapi
import matplotlib.pyplot as plt

candidates = weird['country'].tolist()
name = 'Norway' if 'Norway' in candidates else candidates[0]
row = world_eq[world_eq['country'] == name].iloc[0]
centroid = row.geometry.centroid
rep_point = row.geometry.representative_point()

fig, ax = plt.subplots(figsize=(6, 8))
gpd.GeoSeries([row.geometry], crs=world_eq.crs).plot(ax=ax, color='lightgray', edgecolor='black')
gpd.GeoSeries([centroid], crs=world_eq.crs).plot(ax=ax, color='red', markersize=40)
gpd.GeoSeries([rep_point], crs=world_eq.crs).plot(ax=ax, color='blue', markersize=40)
ax.set_title(f'{name}: centroid vs representative_point')
ax.set_axis_off()
plt.show()

print('Czerwony = centroid')
print('Niebieski = representative_point (zawsze lezy wewnatrz geometrii)')
````

## 7. Zapis warstwy do pliku

### 7.1 QgsVectorFileWriter

Glowna klasa do zapisu warstwy wektorowej do pliku to `QgsVectorFileWriter`.

## 8. Processing Toolbox - uruchamianie algorytmow ze skryptu

Processing Toolbox zawiera ponad 1000 algorytmow geoprzetwarzania:


## 9. Przyklad kompletnego potoku przetwarzania

Ponizszy skrypt laczy wszystkie elementy wykladu w jeden spojny potok.
Wykonuje analize: dla kazdego kraju wyznacza centroid, oblicza pole powierzchni
w km2, a nastepnie zapisuje wyniki do pliku GeoPackage.

```python
# potok_analityczny.py
# Kompletny przyklad laczacy geometrie, reprojekcje i zapis

import processing
from qgis.core import (
    QgsProject, QgsVectorLayer, QgsVectorFileWriter,
    QgsCoordinateReferenceSystem, QgsCoordinateTransform,
    QgsDistanceArea, QgsUnitTypes, QgsFeature, QgsGeometry
)

def potok(nazwa_warstwy_wejsciowej, sciezka_wyjsciowa, kontynent=None):

    # --- Krok 1: wczytaj warstwe wejsciowa ---
    warstwy = QgsProject.instance().mapLayersByName(nazwa_warstwy_wejsciowej)
    if not warstwy:
        print(f'Blad: warstwa "{nazwa_warstwy_wejsciowej}" nie istnieje.')
        return
    wejscie = warstwy[0]
    print(f'Warstwa wejsciowa: {wejscie.name()} ({wejscie.featureCount()} obiektow)')

    # --- Krok 2: opcjonalne filtrowanie po kontynencie ---
    if kontynent:
        wejscie.selectByExpression(f'"continent" = \'{kontynent}\'')
        print(f'Filtr kontynent={kontynent}: {wejscie.selectedFeatureCount()} obiektow')

    # --- Krok 3: reprojekcja do ukladu metrycznego ---
    repr_wynik = processing.run('native:reprojectlayer', {
        'INPUT':      wejscie if not kontynent else
                      QgsProcessingFeatureSourceDefinition(wejscie.id(), True),
        'TARGET_CRS': QgsCoordinateReferenceSystem('EPSG:3035'),
        'OUTPUT':     'TEMPORARY_OUTPUT'
    })
    warstwa_metr = repr_wynik['OUTPUT']
    print(f'Reprojekcja do EPSG:3035 zakonczona.')

    # --- Krok 4: oblicz pole powierzchni i dodaj do atrybutow ---
    warstwa_metr.startEditing()
    from qgis.core import QgsField
    from PyQt5.QtCore import QVariant
    warstwa_metr.addAttribute(QgsField('area_km2', QVariant.Double))
    warstwa_metr.commitChanges()

    idx = warstwa_metr.fields().indexOf('area_km2')
    warstwa_metr.startEditing()
    for obiekt in warstwa_metr.getFeatures():
        pole_m2 = obiekt.geometry().area()
        pole_km2 = round(pole_m2 / 1_000_000, 2)
        warstwa_metr.changeAttributeValue(obiekt.id(), idx, pole_km2)
    warstwa_metr.commitChanges()
    print('Pole powierzchni obliczone.')

    # --- Krok 5: wyznacz centroidy ---
    centr_wynik = processing.run('native:centroids', {
        'INPUT':  warstwa_metr,
        'OUTPUT': 'TEMPORARY_OUTPUT'
    })
    centroidy = centr_wynik['OUTPUT']
    print(f'Centroidy wyznaczone: {centroidy.featureCount()} obiektow.')

    # --- Krok 6: zapisz do pliku ---
    opcje = QgsVectorFileWriter.SaveVectorOptions()
    opcje.driverName   = 'GPKG'
    opcje.fileEncoding = 'UTF-8'
    opcje.layerName    = 'centroidy'

    blad, kom, _, _ = QgsVectorFileWriter.writeAsVectorFormatV3(
        centroidy, sciezka_wyjsciowa,
        QgsProject.instance().transformContext(), opcje
    )

    if blad == QgsVectorFileWriter.NoError:
        print(f'Wynik zapisany: {sciezka_wyjsciowa}')
    else:
        print(f'Blad zapisu: {kom}')

    # --- Krok 7: dodaj wynik do projektu ---
    centroidy.setName('centroidy_wynik')
    QgsProject.instance().addMapLayer(centroidy)
    wejscie.removeSelection()
    iface.mapCanvas().refresh()
    print('Gotowe.')


# Uruchomienie
potok(
    nazwa_warstwy_wejsciowej='kraje',      # zmien na nazwe swojej warstwy
    sciezka_wyjsciowa='C:/GIS_kurs/centroidy_europy.gpkg',
    kontynent='Europe'
)
```


````python
# [JUPYTER] Kompletny potok analityczny w GeoPandas

try:
    import geopandas as gpd
    import matplotlib.pyplot as plt
    import matplotlib.cm as cm
    import numpy as np

    # Krok 1: wczytaj dane
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
    print(f'Krok 1: zaladowano {len(world)} krajow')

    # Krok 2: filtrowanie
    europa = world[world['continent'] == 'Europe'].copy()
    print(f'Krok 2: filtrowanie -> {len(europa)} krajow europejskich')

    # Krok 3: reprojekcja do ETRS89-LAEA (metryczny)
    europa_m = europa.to_crs('EPSG:3035')
    print(f'Krok 3: reprojekcja do EPSG:3035')

    # Krok 4: oblicz pole powierzchni
    europa_m['area_km2'] = (europa_m.geometry.area / 1_000_000).round(2)
    print(f'Krok 4: obliczono pole powierzchni')
    print(europa_m[['name','area_km2']].sort_values('area_km2', ascending=False).head(5).to_string())

    # Krok 5: centroidy
    centroidy = europa_m.copy()
    centroidy['geometry'] = europa_m.geometry.centroid
    print(f'\nKrok 5: wyznaczono {len(centroidy)} centroidow')

    # Krok 6: wizualizacja wynikow
    fig, ax = plt.subplots(figsize=(11, 8))

    europa_m.plot(
        column='area_km2', ax=ax,
        cmap='YlOrRd', legend=True,
        edgecolor='white', linewidth=0.5,
        legend_kwds={'label': 'Pole powierzchni (km2)',
                     'orientation': 'horizontal', 'shrink': 0.6}
    )
    centroidy.plot(ax=ax, color='navy', markersize=5, zorder=5)

    for _, row in centroidy.iterrows():
        ax.annotate(
            row['name'],
            xy=(row.geometry.x, row.geometry.y),
            xytext=(3, 3), textcoords='offset points',
            fontsize=6, color='black'
        )

    ax.set_title('Europa - pole powierzchni i centroidy krajow (EPSG:3035)', fontsize=12)
    ax.axis('off')
    plt.tight_layout()
    plt.show()

    print('\nKrok 6: wizualizacja gotowa.')
    print(f'Potok zakonczony. Przetworzono {len(europa)} krajow.')

except ImportError:
    print('Zainstaluj: pip install geopandas matplotlib numpy')
````

## 10. Podsumowanie i slownik pojec

### 10.1 Co omowilismy na wykladzie 3

- tworzenie geometrii (`QgsGeometry`) z punktow, linii i poligonow,
- odczyt informacji z geometrii: wspolrzedne, lista wierzcholkow, centroid, bounding box,
- podstawowe metody pomiarowe: `.area()`, `.length()`, `.distance()`,
- transformacje pojedynczej geometrii: bufor, uproszczenie, centroid, convex hull,
- operacje na zbiorach geometrii: przeciecie, suma, roznica, roznica symetryczna,
- relacje przestrzenne: `intersects`, `contains`, `within`, `touches`, `overlaps`,
- reprojekcje geometrii przez `QgsCoordinateTransform`,
- generowanie danych testowych programowo: siatka, promienie, ksztalty,
- statystyki i rankingi atrybutow na zywo,
- zaleznosci miedzy panstwami: sasiedzi pierwszego i drugiego stopnia,
- przypadek, w ktorym centroid kraju nie lezy w kraju i kiedy warto uzyc `representative_point()`,
- tworzenie nowej warstwy w pamieci z definiowaniem pol atrybutow,
- zapis warstwy do pliku przez `QgsVectorFileWriter`,
- uruchamianie algorytmow Processing Toolbox przez `processing.run()`,
- wzorzec kompletnego potoku analitycznego.

### 10.2 Slownik kluczowych pojec

| Pojecie | Definicja |
|---|---|
| **Bufor** | Obszar otaczajacy geometrie w zadanej odleglosci |
| **Bounding box** | Najmniejszy prostokat rownolegly do osi, ktory obejmuje geometrie |
| **Centroid** | Geometryczny srodek obiektu; nie zawsze lezy wewnatrz geometrii |
| **Representative point** | Punkt reprezentujacy geometrie, zawsze lezacy wewnatrz niej |
| **Intersects** | Relacja mowiaca, ze geometrie maja czesc wspolna |
| **Touches** | Relacja mowiaca, ze geometrie stykaja sie granica |
| **Reprojekcja** | Zamiana wspolrzednych z jednego CRS do drugiego |
| **CRS metryczny** | Uklad wspolrzednych pozwalajacy liczyc odleglosci i pola w metrach |
| **Atrybut** | Dane opisowe obiektu, np. nazwa, populacja, kraj |
| **Sasiad drugiego stopnia** | Kraj, ktory nie graniczy bezposrednio, ale graniczy z naszym sasiadem |
